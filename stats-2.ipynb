### Q8

moe = z * np.sqrt(0.25 / n)  
np.sqrt(0.25 / n) = moe / z  
0.25 / n = moe**2 / z**2  
n / 0.25 = z**2 / moe**2  
n = (z/moe)**2 / 4  

```python
moe = 0.04
z = 1.96 # for 95% confidence interval
n = (z/moe)**2 / 4
n
```
